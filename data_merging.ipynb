```python
import numpy as np

import pandas as pd 
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
import seaborn as sns
```

```python
crash = pd.read_csv('crashes_columns_eleminated.csv')
casualties = pd.read_csv('casualties_columns_eleminated.csv')
```

```python
merge1=pd.merge(crash,casualties,how='right',on='Crash_Ref_Number')
```

```python
merge1
```

```python
merge1.info()
```

```python
demographics = pd.read_csv('demographics_columns_eleminated.csv')
```

```python
merge2=pd.merge(merge1,demographics,how='right',on='Crash_Ref_Number')
```

```python
merge2.info()
```

```python
cac = pd.read_csv('circumstances_and_conditions_columns_eleminated.csv')
```

```python
cac
```

```python
cac1=cac.rename(columns={'Condition_Ref_Number': 'Crash_Ref_Number'})
```

```python
merge3=pd.merge(merge2,cac1,how='right',on='Crash_Ref_Number')
```

```python
merge3.info()
```

```python
fnf = pd.read_csv('factors_and_features_columns_eleminated.csv')
```

```python
fnf1=fnf.rename(columns={'factors_Ref_Number': 'Crash_Ref_Number'})
```

```python
fnf1
```

```python
merge4=pd.merge(merge3,fnf1,how='right',on='Crash_Ref_Number')
```

```python
merge4
```
